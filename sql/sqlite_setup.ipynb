`conn = sqlite3.connect(":memory:")`

- `conn`: Interact with database
- `sqlite3`: Built-in Python library to work with sqlite databases
- `.connect()`: Create a connection to a database
- `(":memory:")`: Enable sqlite to create a temporary database in RAM and disable once the program ends


`cursor = conn.cursor()`
- `cursor`: Execute SQL statements and fetch results

```python
# Print
cursor.execute("SELECT 'Hello World' AS Message")
print(cursor.fetchall())
```

`cursor.execute("text")`
- `cursor.execute()`: Run SQL

`print(cursor.fetchall())`
- `.fetchall()`: Retrieves all rows from the last executed query
